Let's learn some basic plotting commmands.  
<br>
<br>Import the matplotlib.pyplot package and give it the nickname "plt"

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-30, 30)
```

```python
def mygaussian(x_pts, amplitude, centroid, sigmawidth, offset):
    
    return amplitude* np.exp(  -0.5*(  (centroid - x_pts)/sigmawidth )**2  )  + offset
```

```python
y = mygaussian(x, 2, 15, 3, 5)


plt.figure()
plt.plot( x, y)
plt.show()    # not necessary but good practice
plt.close()   # not necessary but good practice
```

```python
plt.scatter(x, y)
```

```python
plt.errorbar( x, y, xerr=3, yerr=np.random.random( len(y) )*0.5 )
```

```python
plt.figure( dpi=150 )
plt.plot( x, y, 'g--')
plt.scatter(x, y, marker='x', c='red', s=40)
plt.errorbar( x, y, xerr=3, yerr=np.random.random( len(y) )*0.5 )
plt.show()    # not necessary but good practice
plt.close()   # not necessary but good practice
```
